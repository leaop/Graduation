```python
df_raça = pd.read_csv('RacaLabels.csv')
df_cor = pd.read_csv('CorLabels.csv')
```

```python
df.dropna(inplace=True)
```

```python
df.isna().sum() / len(df) * 100
```

```python
df_visu = pd.merge(df, df_raça, left_on='raca1', right_on='BreedID', how='left')
df_visu = pd.merge(df_visu, df_raça, left_on='raca2', right_on='BreedID', how='left')
```

```python
df_visu.rename(columns={'BreedName_x':'Raça_1'}, inplace=True)
df_visu.rename(columns={'BreedName_y':'Raça_2'}, inplace=True)
```

```python
df_visu = pd.merge(df_visu, df_cor, left_on='Cor1', right_on='ColorID', how='left')
df_visu = pd.merge(df_visu, df_cor, left_on='Cor2', right_on='ColorID', how='left')
df_visu = pd.merge(df_visu, df_cor, left_on='Cor3', right_on='ColorID', how='left')
```

```python
df_visu.rename(columns={'ColorName_x':'Cor_1'}, inplace=True)
df_visu.rename(columns={'ColorName_y':'Cor_2'}, inplace=True)
df_visu.rename(columns={'ColorName':'Cor_3'}, inplace=True)
```

```python
df_visu.drop(['raca1', 'raca2', 'Cor1', 'Cor2', 'Cor3', 'BreedID_x', 'Type_x',
              'BreedID_y', 'Type_y', 'ColorID_x', 'ColorID_y', 'ColorID', ], axis=1, inplace=True)
```

```python
map_type = {1:'cachorro', 2:'gato'}
map_gen = {1:'macho', 2:'femea'}
map_TQA = {1:'pequeno', 2:'medio', 3:'grande', 4:'extra grande', 0:'nao especificado'}
map_TPL = {1:'curto', 2:'medio', 3:'longo', 0:'nao especificado'}
map_VAC = {1:'sim', 2:'nao', 3:'incerto'}
map_VER = map_VAC
map_CAS = map_VAC
map_SAU = {1:'saudavel', 2:'pequeno ferimento', 3:'ferimento sério', 0:'nao especificado'}
map_TAX = {0:'sem taxa'}

df_visu['Tipo'] = df_visu['Tipo'].replace(map_type)
df_visu['Genero'] = df_visu['Genero'].replace(map_gen)
df_visu['TamanhoQuandoAdulto'] = df_visu['TamanhoQuandoAdulto'].replace(map_TQA)
df_visu['TamanhoPelagem'] = df_visu['TamanhoPelagem'].replace(map_TPL)
df_visu['Vacinado'] = df_visu['Vacinado'].replace(map_VAC)
df_visu['Vermifugado'] = df_visu['Vermifugado'].replace(map_VER)
df_visu['Castrado'] = df_visu['Castrado'].replace(map_CAS)
df_visu['Saude'] = df_visu['Saude'].replace(map_SAU)
df_visu['Taxa'] = df_visu['Taxa'].replace(map_TAX) 
```

```python
# Dicionário para armazenar os limites de cada coluna
limites = {}

# Itera pelas colunas numéricas do DataFrame
for col in df_visu.select_dtypes(include='number'):
    q1 = df_visu[col].quantile(0.25)
    q3 = df_visu[col].quantile(0.75)
    IQR = q3 - q1
    sup = q3 + 1.5 * IQR
    limites[col] = {'Limite Superior': sup}

# Imprime os limites para cada coluna
for col, valores in limites.items():
    print(f'Limites para a coluna {col}:')
    print('Limite Superior:', valores['Limite Superior'])
    print()
```
