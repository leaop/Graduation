### Exemplos e Exercícios sobre Testes de Hipóteses Estatísticos: Teste F (ANOVA) e Teste do Qui-quadrado (independência)

notações do aluno:
- SQT 
    - primeiro passo soma de todos os elementos/pela quantidade de elementos = média das médias
    - segundo passo: (cada elemento - média das médias)²
    - se quiser a variância: divide o resultado pelos graus de liberdade[(n*m) - 1]
- SQD (soma dos quadrados de dentro)
    - equivale a variação de cada elemento para sua média
    - equivale a fazer um SQT de cada coluna(m)
    - para saber a variancia: m*(n-1)
- SQE (some dos quadrados entre)
    - (a diferença entre a média da coluna com a SQT)²
    - a operação se repete para cada termo(n) da coluna(m): (NM1-SQT)²+(NM2-SQT)²+...

```python
from scipy.stats import f_oneway

# Definir os dados
grupo1 = [3, 2, 1]
grupo2 = [5, 3, 4]
grupo3 = [5, 6, 7]

# Executar ANOVA
f_value, p_value = f_oneway(grupo1, grupo2, grupo3)

# Imprimir resultados
print("Valor de F:", f_value)
print("Valor de p:", p_value)

if p_value < 0.10:
    print("Há diferenças significativas na produção média das plantas entre os três grupos.")
else:
    print("Não há diferenças significativas na produção média das plantas entre os três grupos.")
```

# Teste do qui-quadrado

## Exemplo 1

Suponha que temos os seguintes dados de uma pesquisa sobre preferência de cores de camisetas entre homens e mulheres:

image.png

**A pergunta que queremos responder é se há uma diferença significativa na preferência de cores entre homens e mulheres.** 

* Considere o nível de significância de 0,05

## Passo 1: Definir as hipóteses

As hipóteses são:

* Hipótese nula (H0): não há diferença significativa na preferência de cores entre homens e mulheres.
* Hipótese alternativa (H1): há diferença significativa na preferência de cores entre homens e mulheres.

## Passo 2: Calcular o qui-quadrado

Vamos calcular o qui-quadrado usando a biblioteca SciPy em Python. 

Primeiro, precisamos importar a biblioteca e criar uma matriz de contingência com os dados:

```python
import numpy as np
from scipy.stats import chi2_contingency

# criar a matriz de contingência
obs = np.array([[20, 30, 10, 5], [15, 25, 20, 10]])
print(obs)
```

Agora podemos usar a função chi2_contingency para calcular o qui-quadrado e outras informações relevantes:

```python
chi2, p, dof, expected = chi2_contingency(obs)
```

Onde:

* chi2: o valor do qui-quadrado.
* p: o valor p (probabilidade) associado ao qui-quadrado.
* dof: graus de liberdade.
* expected: matriz de contingência esperada.

```python
expected
```

## Passo 3: Interpretar o resultado

Podemos imprimir o valor do qui-quadrado e o valor p:

```python
print(f"Qui-quadrado: {chi2:.2f}")
print(f"Valor p: {p:.4f}")
print("Graus de liberdade: ", dof)
```

image.png

image.png

O valor do qui-quadrado é 5,99 e o valor p é 0,1120. Como o valor p é maior que o nível de significância de 0,05, não podemos rejeitar a hipótese nula. Portanto, não há evidência de que há uma diferença significativa na preferência de cores entre homens e mulheres.

## Passo 4: Conclusão

Com base no teste do qui-quadrado, não encontramos evidências de que há uma diferença significativa na preferência de cores entre homens e mulheres.

## Exemplo 2
Suponha que um pesquisador esteja interessado em investigar se existe uma associação entre o gênero dos estudantes de uma escola (masculino ou feminino) e sua preferência por diferentes tipos de música (pop, rock, sertanejo). Para isso, o pesquisador selecionou aleatoriamente 100 estudantes da escola e perguntou qual era sua música favorita. Os resultados foram os seguintes:

* 40 estudantes preferem pop (20 meninas e 20 meninos)
* 30 estudantes preferem rock (20 meninas e 10 meninos)
* 30 estudantes preferem sertanejo (10 meninas e 20 meninos)

O pesquisador gostaria de saber se existe uma **associação estatisticamente significativa** entre o gênero dos estudantes e sua preferência por música.

Para realizar essa análise, podemos usar o teste do qui-quadrado de independência. O qui-quadrado é um teste estatístico não paramétrico que avalia se existe uma diferença significativa entre a distribuição observada de dados e uma distribuição esperada.

A hipótese nula do teste é que **não há associação entre as variáveis de interesse (no caso, gênero e preferência musical)**. A **hipótese alternativa é que existe uma associação entre as variáveis**.

A seguir, vamos passo a passo como realizar esse teste:

# Passo 1: Definir as hipóteses

* H0: Não há associação entre gênero e preferência musical
* Ha: Existe associação entre gênero e preferência musical

# Passo 2: Calcular as frequências esperadas

Para calcular as frequências esperadas, podemos usar a seguinte fórmula:

*frequência esperada = (total da linha * total da coluna) / total de observações*

Assim, podemos calcular as frequências esperadas para cada célula da tabela:

```python
import numpy as np
from scipy.stats import chi2_contingency

# Dados da tabela
dados_observado = np.array([[20, 10, 20], [20, 20, 10]])
print(dados_observado)

# Realizando o teste do qui-quadrado
stat, p, dof, expected = chi2_contingency(dados_observado)
```

```python
print("Valores esperados: ")
print(expected)
```

```python
# Imprimindo os resultados
print("Estatística de teste: ", stat)
print(f"p-valor : {p:.4f}")
print("Graus de liberdade: ", dof)
```

image.png

image.png

# Conclusão

O valor do qui-quadrado é 6,66 e o valor p é 0,0357. Como o valor p é menor que o nível de significância de 0,05, *podemos rejeitar a hipótese nula*. Portanto, **há associação entre as variáveis de interesse (no caso, gênero e preferência musical)**. A **hipótese alternativa é que existe uma associação entre as variáveis**.
