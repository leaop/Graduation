As Redes Neurais Convolucionais (CNNs, do inglês Convolutional Neural Networks) são uma classe de Redes Neurais Artificiais profundas, que têm mostrado excelente desempenho em tarefas de visão computacional, como reconhecimento de imagens, detecção de objetos e segmentação semântica. Sua estrutura é inspirada no córtex visual humano, capaz de identificar padrões complexos e hierarquias de características em imagens.

#### Definição de Redes Neurais Convolucionais

Uma CNN é composta por várias camadas que automaticamente e adaptativamente aprendem hierarquias de características a partir de dados de entrada. Essas camadas incluem convoluções que filtram os dados de entrada para criar mapas de características, camadas de pooling que reduzem a dimensão dos dados preservando as características mais importantes, e camadas totalmente conectadas que classificam a entrada com base nas características extraídas.

#### Importância das CNNs na Visão Computacional

CNNs têm sido fundamentais no avanço da visão computacional, superando outros métodos em precisão e eficiência em uma vasta gama de tarefas. Elas são capazes de capturar a estrutura espacial e padrões em imagens de maneira eficaz, o que as torna particularmente adequadas para tarefas que envolvem análise de imagens. Seu sucesso em competições e aplicações práticas tem impulsionado a adoção e desenvolvimento de técnicas de aprendizado profundo em visão computacional.

#### Diferença entre CNNs e Redes Neurais Artificiais (RNAs) Tradicionais

Enquanto as RNAs tradicionais, como o Perceptron Multicamadas (MLP), consistem em camadas totalmente conectadas onde cada neurônio de uma camada está conectado a todos os neurônios da próxima camada, as CNNs introduzem o conceito de camadas convolucionais. Essas camadas utilizam um conjunto de filtros compartilhados que varrem a entrada para detectar padrões espaciais. Diferentemente das RNAs, que não consideram a estrutura espacial dos dados de entrada, as CNNs preservam as relações espaciais entre pixels, o que as torna mais eficientes para processamento de imagens e dados com dimensões similares. Além disso, o compartilhamento de pesos nas camadas convolucionais reduz significativamente o número de parâmetros a serem aprendidos, aumentando a eficiência computacional e diminuindo o risco de sobreajuste.


### Componentes de uma Rede Neural Convolucional (CNN)

- Camadas convolucionais: função e operação.
- Funções de ativação em CNNs: ReLU e suas variantes.
- Pooling: Max Pooling e Average Pooling.
- Camadas Fully Connected (Dense): integração à arquitetura CNN.

Uma CNN é composta por diversas camadas especializadas que trabalham em conjunto para extrair padrões e características importantes das imagens, permitindo a realização de tarefas complexas de visão computacional. Cada tipo de camada tem uma função específica e contribui de maneira única para o processo de aprendizado da rede.

#### Camadas Convolucionais: Função e Operação

- **Função**: As camadas convolucionais são o coração de uma CNN. Elas aplicam um conjunto de filtros à imagem de entrada para produzir mapas de características, capturando padrões espaciais como bordas, texturas e formas. 
- **Operação**: Durante a passagem para frente (forward pass), cada filtro convolucional desliza (ou convoluciona) sobre a imagem de entrada, realizando uma multiplicação elemento a elemento seguida de uma soma, produzindo assim um mapa de características. Este processo é repetido para diversos filtros para capturar uma ampla gama de características.

#### Funções de Ativação em CNNs: ReLU e suas Variantes

- **ReLU (Rectified Linear Unit)**: A função de ativação ReLU é amplamente usada em CNNs devido à sua simplicidade e eficiência. Ela introduz não-linearidade no modelo, permitindo que a rede aprenda funções mais complexas. A ReLU é definida como f(x) = max(0, x), zerando todos os valores negativos de seu input.
- **Variantes**: Existem várias variantes da ReLU projetadas para melhorar o desempenho e a convergência da rede, como a Leaky ReLU, a Parametric ReLU (PReLU) e a Exponential Linear Unit (ELU). Essas variantes tentam resolver o problema dos "neurônios mortos", que ocorre quando um grande número de ativações se torna zero e não contribui para o ajuste do modelo.

#### Pooling: Max Pooling e Average Pooling

- **Max Pooling**: Esta camada reduz as dimensões espaciais (altura e largura) do mapa de características, mantendo apenas o valor máximo de uma região definida. Isso ajuda a tornar a representação mais compacta e invariante a pequenas translações na imagem.
- **Average Pooling**: Similar ao Max Pooling, mas, em vez de selecionar o valor máximo, calcula a média dos valores em uma região específica do mapa de características. Embora menos comum que o Max Pooling, o Average Pooling pode suavizar as características extraídas.

#### Camadas Fully Connected (Dense): Integração à Arquitetura CNN

- **Função**: Após a extração de características e redução de dimensionalidade pelas camadas convolucionais e de pooling, as camadas totalmente conectadas (dense) combinam essas características para realizar a classificação ou regressão. Essas camadas são semelhantes às encontradas em redes neurais tradicionais, onde cada neurônio está conectado a todos os neurônios da camada anterior.
- **Operação**: As camadas totalmente conectadas aprendem combinações não-lineares dos recursos de alto nível representados nos mapas de características, essenciais para realizar tarefas específicas como classificação de imagens. A última camada totalmente conectada geralmente tem um número de neurônios igual ao número de classes de saída e é seguida por uma função de ativação softmax para classificação multiclasse.

Esses componentes trabalham em conjunto para permitir que as CNNs aprendam representações ricas e hierárquicas de imagens, tornando-as ferramentas poderosas para visão computacional e outras aplicações que envolvem dados de entrada em forma de grade, como séries temporais e dados espaciais.

```python
# Realiza a leitura da imagem
img = cv2.imread("casac.png")

# Converte a imagem de BGR para RGB para exibição correta com matplotlib
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Exibe a imagem original
plt.figure(figsize=(3, 3))
plt.imshow(img_rgb)
plt.title('Imagem Original')
plt.show()

# Converte a imagem para preto e branco
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Exibe a imagem em preto e branco
plt.figure(figsize=(3, 3))
plt.imshow(img_gray, cmap='gray')
plt.title('Imagem Preto e Branco')
plt.show()

# Dimensões da imagem
print('A imagem tem dimensão',img.shape[0],'x',img.shape[1],'pixels')
```

# Realizar a Convolução

```python
def filtro_cnn(img, filtro):
    # Cria uma matriz da dimensão da imagem mas só com valores nulos
    imgOut = np.zeros_like(img)

    A_IN, L_IN = img.shape
    A_OUT, L_OUT = imgOut.shape

    # Recebe a dimensão do filtro
    A, L = filtro.shape

    # Stride
    S = 1

    # Itera sobre as linhas e colunas
    for i in range(A_OUT - A + 1):
        for j in range(L_OUT - L + 1):
            # Convolução
            mult = np.multiply(img_gray[i*S: i*S + A, j*S: j*S + L], filtro)
            # Soma
            soma = np.sum(mult)
            # Salva o valor da soma no mapa de recursos
            imgOut[i, j] = soma

    # Ajusta os valores de imgOut para o intervalo 0-255
    imgOut = cv2.normalize(imgOut, None, 0, 255, cv2.NORM_MINMAX)

    # Exibe a imagem após aplicar o filtro
    plt.figure(figsize=(3,3))
    plt.imshow(imgOut, cmap='gray')
    plt.title('Saida do Neurônio')
    plt.show()
```

```python
filtro_linhas_horizontais = np.array([[-1, -1, -1],
                                    [0, 0, 0],
                                    [1, 1, 1]])
filtro_cnn(img_gray, filtro_linhas_horizontais)
```

```python
filtro_linhas_verticais = np.array([[-1, 0, 1],
                                      [-1, 0, 1],
                                      [-1, 0, 1]])
filtro_cnn(img_gray, filtro_linhas_verticais)
```

```python
filtro_linhas_horizontais_e_verticais = np.array([[0, -1, 0],
                                                  [-1, 4, -1],
                                                  [0, -1, 0]])
filtro_cnn(img_gray, filtro_linhas_horizontais_e_verticais)
```

```python
filtro_diagonais_esquerdas = np.array([[-1, -1, 2],
                                       [-1, 2, -1],
                                       [2, -1, -1]])

filtro_cnn(img_gray, filtro_diagonais_esquerdas)
```

A evolução das arquiteturas de Redes Neurais Convolucionais (CNNs) tem sido fundamental para o progresso do campo da visão computacional. Cada arquitetura introduziu novos conceitos e técnicas, superando os desafios encontrados por suas predecessoras. Esta seção explora algumas das arquiteturas mais influentes e como elas contribuíram para o avanço da análise de imagens.

#### LeNet-5: A Primeira CNN

- **Introdução**: Desenvolvida por Yann LeCun et al. em 1998, LeNet-5 é considerada a primeira CNN de sucesso, projetada para reconhecimento de dígitos manuscritos.
- **Estrutura**: Composta por camadas convolucionais alternadas com camadas de subsampling (pooling), seguidas de camadas totalmente conectadas.
- **Impacto**: Demonstrou a viabilidade das CNNs em tarefas de visão computacional, estabelecendo a base para futuras pesquisas na área.

#### AlexNet: O Avanço para Redes Profundas

- **Introdução**: Apresentada por Alex Krizhevsky, Ilya Sutskever e Geoffrey Hinton em 2012, AlexNet marcou o renascimento do interesse por redes neurais profundas ao vencer o desafio ImageNet LSVRC-2012.
- **Estrutura**: Possui uma arquitetura mais profunda e ampla que a LeNet-5, com camadas convolucionais seguidas de camadas de pooling e totalmente conectadas, e introduziu o uso da função de ativação ReLU.
- **Impacto**: Demonstrou a eficácia das CNNs em conjuntos de dados de imagem de grande escala, inspirando o desenvolvimento de arquiteturas mais profundas e complexas.

#### VGGNet: Padronização da Profundidade

- **Introdução**: Desenvolvida por Karen Simonyan e Andrew Zisserman da Universidade de Oxford em 2014, a VGGNet é conhecida por sua arquitetura simples e uniforme.
- **Estrutura**: Caracteriza-se por sua profundidade, com até 19 camadas, e o uso exclusivo de filtros convolucionais de tamanho 3x3.
- **Impacto**: Estabeleceu a norma para a configuração de CNNs, mostrando que a profundidade é crucial para alcançar um bom desempenho.

#### GoogLeNet (Inception): Inovação em Módulos de Convolução

- **Introdução**: Introduzida em 2014, a GoogLeNet, ou Inception v1, introduziu o conceito de "módulo inception", visando a eficiência computacional.
- **Estrutura**: Incorpora módulos inception, que são blocos de camadas paralelas com diferentes tamanhos de filtros, permitindo que a rede capture informações em várias escalas.
- **Impacto**: Aumentou significativamente a profundidade e a largura da rede sem um aumento proporcional nos parâmetros, melhorando a eficiência e o desempenho.

#### ResNet: Introdução de Conexões Residuais

- **Introdução**: Proposta por Kaiming He et al. em 2015, a ResNet introduziu o conceito de conexões residuais para facilitar o treinamento de redes muito profundas.
- **Estrutura**: As conexões residuais permitem que as ativações "saltem" camadas, enfrentando o problema do desvanecimento do gradiente ao treinar redes profundas.
- **Impacto**: Permitiu o desenvolvimento de redes com mais de 100 camadas, alcançando desempenho notável em várias tarefas de visão computacional.

#### Comparação entre as Arquiteturas: Performance, Complexidade e Uso Prático

- **Performance**: Cada arquitetura trouxe melhorias significativas em precisão e eficiência, com ResNet estabelecendo novos padrões para o desempenho em classificação de imagens.
- **Complexidade**: Enquanto LeNet-5 e AlexNet são relativamente simples e diretas, VGGNet, GoogLeNet e ResNet introduziram uma complexidade crescente em termos de profundidade e mecanismos de conexão.
- **Uso Prático**: Apesar de arquiteturas mais profundas e complexas oferecerem melhor desempenho, elas também requerem mais recursos computacionais. A escolha da arquitetura depende do equilíbrio entre precisão necessária e recursos disponíveis.

Estas arquiteturas representam apenas uma fração das inovações em CNNs, cada uma contribuindo para a nossa compreensão e capacidade de processar visualmente dados complexos de maneira eficiente.

```python
# Importação das bibliotecas necessárias
import tensorflow as tf
from tensorflow.keras import layers, models
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.pyplot as plt

def exibir_imagem(train_images, labels, position):
    class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer',
               'dog', 'frog', 'horse', 'ship', 'truck']
    """
    Exibe uma imagem e seu respectivo rótulo (label) de um dataset.

    Parâmetros:
    - train_images: numpy array de imagens.
    - labels: lista ou numpy array de rótulos.
    - position: posição (índice) da imagem a ser exibida no dataset.
    """
    label_index = labels[position][0]
    
    label_name = class_names[label_index]

    # Define o título da figura com o rótulo da imagem
    plt.title(f'Classe: {label_name}')
    # Exibe a imagem
    plt.imshow(train_images[position])
    # Remove os eixos para uma visualização mais clara
    plt.axis('off')
    # Mostra a figura
    plt.show()

# Preparação do Ambiente
## Verificando se o TensorFlow está instalado
print(tf.__version__)
```

```python
# Carregando os dados do arquivo NPZ
data = np.load('cifar10_dataset.npz')

train_images = data['train_images']
train_labels = data['train_labels']
test_images = data['test_images']
test_labels = data['test_labels']
```

```python
exibir_imagem(train_images, train_labels, 2)
```

```python
exibir_imagem(train_images, train_labels, 8)
```

```python
exibir_imagem(train_images, train_labels, 4)
```

```python
## Normalizando os dados para o intervalo [0, 1]
train_images, test_images = train_images / 255.0, test_images / 255.0

## Convertendo os rótulos para o formato one-hot encoding
train_labels = tf.keras.utils.to_categorical(train_labels, 10)
test_labels = tf.keras.utils.to_categorical(test_labels, 10)
```

```python
# Construção de uma CNN Simples para Classificação de Imagens
model = models.Sequential()
# Camada de entrada e primeira camada convolucional
model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)))
model.add(layers.MaxPooling2D((2, 2)))
# Segunda camada convolucional
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
# Terceira camada convolucional
model.add(layers.Conv2D(64, (3, 3), activation='relu'))

# Adicionando camadas totalmente conectadas
model.add(layers.Flatten())
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(10, activation='softmax'))

# Resumo do modelo
model.summary()
```

```python
# Treinamento e Avaliação do Modelo
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Treinamento do modelo
history = model.fit(train_images, train_labels, epochs=3, 
                    validation_data=(test_images, test_labels))
```

```python
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns

# Nomes das classes para CIFAR-10
class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']


# Avaliação do modelo para obter a precisão de teste
test_loss, test_acc = model.evaluate(test_images, test_labels, verbose=2)
print('\nTest accuracy:', test_acc)

# Obtendo as previsões do modelo no conjunto de teste
predictions = model.predict(test_images)
predicted_classes = np.argmax(predictions, axis=1)
true_classes = np.argmax(test_labels, axis=1)

# Gerando o relatório de classificação
print("\nRelatório de Classificação:\n")
print(classification_report(true_classes, predicted_classes, target_names=class_names))

# Gerando e visualizando a matriz de confusão
confusion_mtx = confusion_matrix(true_classes, predicted_classes)
plt.figure(figsize=(10, 8))
sns.heatmap(confusion_mtx, annot=True, fmt='d', cmap="Blues", xticklabels=class_names, yticklabels=class_names)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Matriz de Confusão')
plt.show()

# Visualizando os dados de desempenho (history)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()

plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

### Técnicas de Pré-processamento de Imagens para CNNs

- Normalização e redimensionamento.
- Aumento de dados (Data Augmentation): rotação, translação, zoom e flip horizontal.

O pré-processamento de imagens é uma etapa crucial no desenvolvimento de modelos de Redes Neurais Convolucionais (CNNs), essencial para garantir que as imagens estejam no formato correto e otimizado para treinamento. Este processo inclui normalização, redimensionamento e técnicas de aumento de dados (data augmentation), cada uma contribuindo para a melhoria do desempenho e eficiência da rede.

#### Normalização e Redimensionamento

- **Normalização**: Esta técnica ajusta os valores dos pixels das imagens para um intervalo comum, como [0, 1] ou [-1, 1], através da divisão de cada pixel pelo valor máximo (255). A normalização ajuda na estabilização do treinamento da rede, facilitando a sua convergência, reduzindo a variação entre os valores dos pixels e permitindo que a rede aprenda as características importantes das imagens de maneira mais eficiente.

- **Redimensionamento**: Modifica o tamanho das imagens para que todas tenham as mesmas dimensões, um requisito para o processamento pela CNN. O redimensionamento é realizado através de técnicas como interpolação, que ajusta as imagens ao tamanho desejado, mantendo a proporção ou adaptando-as ao novo tamanho sem distorções significativas.

#### Aumento de Dados (Data Augmentation)

O aumento de dados é uma técnica chave para enriquecer o conjunto de dados de treinamento, introduzindo variações artificiais nas imagens, o que pode ajudar a melhorar a robustez e o desempenho do modelo.

- **Rotação**: Gira as imagens por um ângulo definido, criando variações na orientação dos objetos contidos nas imagens, o que é fundamental para ensinar a rede a reconhecer objetos independentemente de sua orientação.

- **Translação**: Move as imagens vertical ou horizontalmente, alterando a posição dos objetos dentro das imagens. Isso ajuda a rede a aprender que a posição do objeto não muda sua natureza ou classe.

- **Zoom**: Aplica um zoom para aumentar ou diminuir as imagens, simulando objetos que estão mais próximos ou mais distantes da câmera. Isso treina a rede para identificar objetos em várias escalas.

- **Flip Horizontal**: Espelha as imagens horizontalmente, gerando uma versão espelhada de cada imagem. Isso aumenta a diversidade do conjunto de dados, ensinando a rede a reconhecer objetos, independentemente de sua orientação horizontal.


A aplicação cuidadosa dessas técnicas de pré-processamento e aumento de dados não só prepara as imagens para uma ingestão mais eficiente pela rede como também amplia a diversidade do conjunto de dados de treinamento. Isso é crucial para construir modelos de CNN que são capazes de generalizar bem e performar de maneira robusta em dados novos e não vistos, reduzindo o risco de sobreajuste e melhorando a precisão da classificação.


```python
from tensorflow.keras import datasets
import matplotlib.pyplot as plt

# Carregando o dataset Fashion MNIST
(train_images, train_labels), (test_images, test_labels) = datasets.fashion_mnist.load_data()
```

### Ferramentas e Bibliotecas para Modelagem de CNNs

- Uso de TensorFlow e Keras para modelagem de CNNs.
- Utilização de datasets públicos para treinamento e teste.

No desenvolvimento de Redes Neurais Convolucionais (CNNs), o uso de ferramentas e bibliotecas adequadas é fundamental para simplificar o processo de implementação, treinamento e avaliação de modelos. Entre as mais populares e poderosas para trabalhar com CNNs estão TensorFlow e Keras, além de várias fontes de datasets públicos que servem para treinar e testar modelos.

#### Uso de TensorFlow e Keras

- **TensorFlow**: Desenvolvido pelo Google Brain, TensorFlow é uma biblioteca de software de código aberto para computação numérica que utiliza gráficos de fluxo de dados. É amplamente utilizado para pesquisa e produção em aprendizado de máquina, particularmente em redes neurais profundas. TensorFlow oferece uma variedade de ferramentas e bibliotecas que permitem desenvolver e treinar modelos de forma eficiente.

- **Keras**: Keras é uma interface de alto nível para redes neurais, escrita em Python e capaz de rodar em cima do TensorFlow, CNTK ou Theano. Foi desenvolvida com o objetivo de permitir uma experimentação rápida. Keras é conhecida por sua facilidade de uso e por permitir que tanto pesquisadores quanto desenvolvedores possam construir protótipos de redes neurais e experimentar com elas de forma simples e rápida.

#### Características do TensorFlow e Keras

- **Flexibilidade e Escalabilidade com TensorFlow**: O TensorFlow oferece flexibilidade para experimentar com arquiteturas de CNNs avançadas, suportando operações complexas e permitindo escalar o treinamento para múltiplas GPUs e dispositivos distribuídos se necessário.
  
- **Simplicidade e Rapidez com Keras**: Keras simplifica o processo de construção de uma rede neural. Com apenas algumas linhas de código, é possível montar uma CNN, configurar suas camadas, definir funções de ativação e compilar o modelo, tornando o processo de desenvolvimento mais rápido e acessível.

#### Utilização de Datasets Públicos para Treinamento e Teste

A utilização de datasets públicos é essencial para treinar e validar modelos de CNNs. Esses conjuntos de dados, muitas vezes extensos e variados, fornecem uma base rica para o treinamento de modelos robustos e generalizáveis.

- **Datasets Populares**: Alguns dos conjuntos de dados públicos mais utilizados na modelagem de CNNs incluem MNIST (dígitos manuscritos), CIFAR-10 e CIFAR-100 (imagens de objetos em diversas categorias), e ImageNet (milhões de imagens em milhares de categorias).
