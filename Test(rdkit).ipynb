```python
import pandas as pd
```

```python
from rdkit import Chem
from rdkit.Chem.Descriptors import ExactMolWt
from rdkit.Chem.Crippen import MolLogP
from rdkit.Chem.Lipinski import RingCount
```

```python
from rdkit.Chem import PandasTools 
```

```python
SDFFile = '/content/ymdb.sdf'
BRDLigs = PandasTools.LoadSDF(SDFFile)
```

```python
BRDLigs
```

```python
BRDLigs.info()
```

```python
BRDLigs['NumHeavyAtoms']=BRDLigs.apply(lambda x: x['ROMol'].GetNumHeavyAtoms(), axis=1)
```

```python
BRDLigs = BRDLigs[['DATABASE_ID','SMILES','FORMULA','EXACT_MASS','JCHEM_ACCEPTOR_COUNT',
'JCHEM_AVERAGE_POLARIZABILITY','JCHEM_BIOAVAILABILITY','JCHEM_DONOR_COUNT',
'JCHEM_FORMAL_CHARGE','JCHEM_GHOSE_FILTER','JCHEM_IUPAC','ALOGPS_LOGP',
'JCHEM_LOGP','ALOGPS_LOGS','JCHEM_MDDR_LIKE_RULE','JCHEM_NUMBER_OF_RINGS',
'JCHEM_PHYSIOLOGICAL_CHARGE','JCHEM_PKA','JCHEM_PKA_STRONGEST_ACIDIC',
'JCHEM_PKA_STRONGEST_BASIC','JCHEM_POLAR_SURFACE_AREA','JCHEM_REFRACTIVITY',
'JCHEM_ROTATABLE_BOND_COUNT','JCHEM_RULE_OF_FIVE','ALOGPS_SOLUBILITY',
'JCHEM_TRADITIONAL_IUPAC','JCHEM_VEBER_RULE','GENERIC_NAME','SYNONYMS',
'ROMol','JCHEM_EXACT_MASS','JCHEM_FORMULA','JCHEM_MASS','JCHEM_ATOM_COUNT',]]
```

```python
BRDLigs.head()
```

```python
ExactMolWt(Chem.MolFromSmiles('N[C@@H](CCC(N)=O)C(O)=O'))
```

```python
logp = []
for i in BRDLigs.index:
  logp.append(MolLogP(BRDLigs['ROMol'][i]))
```

```python
logp
```

```python
logp = pd.Series(logp)
BRDLigs['LogP'] = logp.values
```

```python
BRDLigs.head()
```

```python
BRDLigs_new = BRDLigs.drop(columns=['JCHEM_AVERAGE_POLARIZABILITY', 'JCHEM_BIOAVAILABILITY', 'JCHEM_FORMAL_CHARGE', 'JCHEM_GHOSE_FILTER',
       'JCHEM_IUPAC', 'ALOGPS_LOGP', 'JCHEM_LOGP', 'ALOGPS_LOGS',
       'JCHEM_MDDR_LIKE_RULE', 'JCHEM_PHYSIOLOGICAL_CHARGE', 'JCHEM_PKA', 'JCHEM_PKA_STRONGEST_ACIDIC',
       'JCHEM_PKA_STRONGEST_BASIC',
       'JCHEM_REFRACTIVITY', 'JCHEM_RULE_OF_FIVE', 'ALOGPS_SOLUBILITY', 'JCHEM_TRADITIONAL_IUPAC',
       'JCHEM_VEBER_RULE','SYNONYMS',
       'JCHEM_EXACT_MASS', 'JCHEM_FORMULA', 'JCHEM_MASS', 'JCHEM_ATOM_COUNT'])
```
